```python
import pandas as pd
import re
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data/full_data.csv")
```

```python
url_set = set({})
def func(s: str, url_set: set):
    if type(s) != str:
        return s
    fd = re.findall('"(.+?)"', s)
    url_set.update(fd)

no = df['entities.urls'].apply(func, args=(url_set, ))
url_set
```

```python
url_ls = list(url_set)
url_srs = pd.Series(url_ls)
url_srs.to_csv("data/entities_urls.csv", index=False)
url_srs
```
